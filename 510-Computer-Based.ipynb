# Random Sample from 2D-Gaussian Distribution

```python
#1: (a-f)

plt.rcParams['figure.figsize'] = 8,8

def generate_and_plot(kx, mu):
    distr = multivariate_normal(
        cov = Kx, mean = mu, 
        seed = 1000)
    data = distr.rvs(size = 5000)
    plt.grid()
    
    plt.plot(data[:, 0], data[:, 1], 'o', c='lime', markeredgewidth = 0.5, markeredgecolor = 'black')
    
    plt.title(r'Random Sample from 2D-Gaussain distribution')
    plt.xlabel(r'$x_1$')
    plt.ylabel(r'$x_2$')
    plt.axis('equal')
    
    # 1b: calculate the principal axes and their lengths
    eigen_val, eigen_vec = np.linalg.eig(Kx)
    principal_axes = eigen_vec
    length = np.sqrt(eigen_val)

    print("principal_axes", principal_axes)
    print("length", length)
    
def generate_elipsoid(kx, mu):
    distr = multivariate_normal(
        cov = Kx, mean = mu, 
        seed = 1000)
    data = distr.rvs(size = 5000)
    plt.grid()
    
    #creating and sketching the Ellipsoid 
    create_ellipse = Ellipse(mu, width=2 * length[0], height=2 * length[1], angle=np.degrees(np.arctan2(principal_axes[1, 0], principal_axes[0, 0])), fill=True, color='blue', label='Ellipsoid Sketch')
    plt.gca().add_patch(create_ellipse)
    
    plt.legend()
    plt.axis('equal')
    plt.grid(True)
    plt.show()
    
#define the mean and covariance matrix 
Kx = np.array([[2.0,1.0], [1.0, 4.0]])
mu = np.array([0,0])
random_seed = 10

generate_and_plot(Kx, mu)
generate_elipsoid(Kx, mu)

#Q2: 
# matrix A
k_vals = 10000
matrix_a = np.array([[1/2, 1/4],[1/2, 3/4]])

# eigenvalues and eigenvectors
eigen_val, eigen_vec = np.linalg.eig(matrix_a)

# q matrix 
matrix_q = eigen_vec

# diagonal matrix 
diagonal_matrix = np.diag(eigen_val)

# k^1000
output = np.diag(eigen_val ** k_vals)
final_output = np.dot(np.dot(Q, output), np.linalg.inv(Q))

# print the output
print(f"Question 2 Output: Print A^K1000 =", final_output)
```
